```python
import nltk

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.text import Text

from string import punctuation
punctuation = list(punctuation)

nltk.download('punkt_tab')
nltk.download('stopwords')

transcript = open('Desc_1.txt', encoding="utf-8").read().lower()


#transcript_sent = sent_tokenize(transcript)
#transcript_sent

transcript_word = word_tokenize(transcript)
transcript_word


stop_words = stopwords.words('english')

filtered_transcript_word = []
for word in transcript_word:
    if word not in stop_words and word not in punctuation:
      filtered_transcript_word.append(word)


from nltk import FreqDist
transcript_fdist = FreqDist(filtered_transcript_word)
transcript_fdist.most_common(50)

text_list = Text(transcript_word)
text_list.concordance("thermostat", lines=50)



```
